```python
drug = df['drug']
coformer = df['sequence']
properties = ['unobstructed']

classification = Classifier()

def predict_property(row, prop):
    return classification.predict_properties(row['drug'], row['sequence'], [prop])[prop]

# Применяем функцию к каждому ряду DataFrame и сохраняем результат в новом столбце 'properties'
df['unobstructed'] = df.apply(lambda row: predict_property(row, 'unobstructed'), axis=1)
df['h_bond_bridging'] = df.apply(lambda row: predict_property(row, 'h_bond_bridging'), axis=1)
df['orthogonal_planes'] = df.apply(lambda row: predict_property(row, 'orthogonal_planes'), axis=1)
```
